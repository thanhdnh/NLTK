```python
from sre_parse import CATEGORIES
from nltk.corpus import brown
brown_tagged_sents= brown.tagged_sents(categories='news')
fd= nltk.FreqDist(brown.words(categories ='news'))
cfd=nltk.ConditionalFreqDist(brown.tagged_words(categories ='news'))
most_freq_words = fd.most_common(100)
likely_tags = dict((word,cfd[word].max())for (word,__)in most_freq_words)
baseline_tagger=nltk.UnigramTagger(model = likely_tags)
baseline_tagger.evaluate(brown_tagged_sents)
```

```python
text = gutenberg.words('austen-persuasion.txt')

brown_sents = brown.sents(categories='news') # tch item
brown_tagged_sents = brown.tagged_sents(categories='news')
affix_tagger = nltk.AffixTagger(train = brown_tagged_sents, affix_length=1, min_stem_length=3) #train data
print(affix_tagger.tag(text)) #test data

```

```python
brown_tagged_sents= brown.tagged_sents(categories='news')
size = int(len(brown_tagged_sents)*0.7)
train_set = brown_tagged_sents[:size]
test_set = brown_tagged_sents[size:]

bigram = nltk.BigramTagger(train_set)
print(bigram.accuracy(train_set))
print(bigram.accuracy(test_set))
```
